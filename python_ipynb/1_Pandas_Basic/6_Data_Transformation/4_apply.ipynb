##### df.apply(function, axis=0)       
##### axis=0 (default)
##### row or column     “Work one row or one column at a time”


```python
df.fillna(0).apply(lambda x: x['Salary']+ x['bonus'],axis=1)
# Runs a function on each row or each column.
```

```python
list1=[]
for i in df['bonus']:
    list1.append(i)
print(list1)
```

```python
for i,x in zip(df['bonus'],df['Salary']):
    print(i+x)
```

##### df.map(function) works on series

```python
df['Salary'].map(lambda x: x+1000)
```

```python
df[['Salary','bonus']].applymap(lambda x: x*2)
```
