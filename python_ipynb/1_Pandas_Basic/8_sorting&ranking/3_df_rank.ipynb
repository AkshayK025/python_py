```python
import pandas as pd

df = pd.read_csv("employees.csv")
df.head()
```

##### Rank salaries (highest = rank 1)

```python
df['salary_rank']=df['salary'].rank(ascending=False)
df.head()
```

##### rank(method ='average')

```python
df["rank_average"] = df["experience"].rank(method="average")
df["rank_min"] = df["experience"].rank(method="min")
df["rank_dense"] = df["experience"].rank(method="dense")

df
```
