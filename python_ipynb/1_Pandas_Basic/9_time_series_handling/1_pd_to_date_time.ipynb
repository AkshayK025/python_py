```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
df['date'].dtype
# data type of date column is object
```

##### pd.to_datetime to coversion another datatype into date datatype

```python
df['date']= pd.to_datetime(df['date'])
print(type(df['date'].dtype))
```

```python
df.date.dt.year
```

```python
df['date'].dt.day
```

```python
print(df.date.dt.month  ) # month number
print(df.date.dt.day    ) # day number
print(df.date.dt.weekday) # day of week (0 = Monday)
```
