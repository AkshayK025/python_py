```python
import pandas as pd

df1 = pd.read_csv("file1.csv")
df2 = pd.read_csv("file2.csv")

print(df1)
print(df2)
```

##### pd.merge() â€“ SQL-style join (most common)

```python
pd.merge(df1, df2, on="id")
```

##### left join

```python
pd.merge(df1, df2, on="id", how="left")
```

##### right join

```python
pd.merge(df1, df2, on="id", how="right")
```

##### ðŸ”¹ Outer Join

```python
result = pd.merge(df1, df2, on="id", how="outer")
print(result)
# Use merge() when you have a common column/key
```
