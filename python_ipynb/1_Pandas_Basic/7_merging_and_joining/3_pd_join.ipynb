```python
import pandas as pd

df1 = pd.read_csv("file1.csv")
df2 = pd.read_csv("file2.csv")

print(df1)
print(df2)
```

```python
# First, set id as index:
df1_indexed = df1.set_index("id")
df2_indexed = df2.set_index("id")
```

#### Join

```python
result = df1_indexed.join(df2_indexed)
print(result)
```

##### ðŸ“Œ join() works best when joining on indexes

```python
df1_indexed.join(df2_indexed, how="left")
```

```python
df1_indexed.join(df2_indexed, how="right")
```

```python
df1_indexed.join(df2_indexed, how="outer")
```
